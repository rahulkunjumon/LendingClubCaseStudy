```python
#Loading the python Libraries to work on the given dataset
import pandas as pd # To work with dataset
import numpy as np # Math library
import seaborn as sns # Graph library that use matplot in background
import matplotlib.pyplot as plt # Plotting the params in seaborn
```

```python
## Import the Data
### We import the data and try to understand the columns and validate its effectiveness to deduce the conclusions
```

```python
#Import the loan.csv file via read_csv
loan_data = pd.read_csv('loan.csv',index_col=0,low_memory=False)
loan_data.head
```

```python
loan_data.shape
loan_data.describe()
data_top = loan_data.head() 
print(data_top)
```

```python
# Data Cleansing
### After understanding the column terminologies, we start with cleaning the data to reduce the clutter and effectively deduce the conclusions
```

```python
print ("Data Cleansing \n" )
print ("Columns in Loan excel which are null " )
null_cols = len(loan_data.columns[loan_data.isna().any()].tolist())
print(null_cols)


#print(loan_data.columns[loan_data.isna().any()].tolist())
result = loan_data.isna().mean()
loan_data = loan_data.loc[:,result < .05]

print ("\nTotal Columns in Loan excel" )
print(loan_data.shape[1])

# dropping the rows having NaN values
loan_data = loan_data.dropna()

print ("\nTotal rowss in Loan excel" )
print(loan_data.shape[0])
print ("\n % of Null data ")  
round(len(loan_data.columns[loan_data.isna().any()].tolist())/loan_data.shape[1] * 100 ,2)

#print(loan_data['loan_status'])
```

```python
###Lets delete the null coumns so that we get a dataset for analysis 
thresh = len(loan_data) * .05
loan_data.dropna(thresh = thresh, axis = 1, inplace = True)

print ("\nTotal Columns in Loan excel" )
print(loan_data.shape[1])
#print(loan_data)
#print(loan_data['loan_status'])
```

```python
## Now we try to identify unique values in a column. This wll help us in cleaning those columns with least unique values(eg. 1) as they wont help in analysis much
loan_data.nunique().sort_values()
```

```python
## Dropping the columns having just 1 unique Value
dropColumns = ['tax_liens','delinq_amnt','chargeoff_within_12_mths','acc_now_delinq','application_type','policy_code','pymnt_plan','initial_list_status','collections_12_mths_ex_med']
loan_data= loan_data.drop(dropColumns,axis=1)
```

```python
## Copy the filtered Data to a new CSV File
loan_data.to_csv('filtered_loan.csv', encoding='utf-8', index=False)
filtered_loan = loan_data[:]
```

```python
## Filtered list shape
filtered_loan.shape
```

```python
## As a part of Data cleansing, Now we need to perform Precision Analysis
### As a part of this we would be performing below Operations
### Convert decimals to nearest 2 places
### Convert dates to proper format as well as seperate them to columns
### Strip the strings from Data which can be categorised
```

```python
## Convert decimals to nearest 2 places
## Notably :
### total_pymnt
### total_pymnt_inv
### total_rec_prncp
### total_rec_int
### last_pymnt_amnt
### annual_inc
filtered_loan.total_pymnt= round(filtered_loan.total_pymnt,2)
filtered_loan.total_pymnt_inv= round(filtered_loan.total_pymnt_inv,2)
filtered_loan.total_rec_prncp= round(filtered_loan.total_rec_prncp,2)
filtered_loan.total_rec_int= round(filtered_loan.total_rec_int,2)
filtered_loan.last_pymnt_amnt= round(filtered_loan.last_pymnt_amnt,2)
filtered_loan.annual_inc= round(filtered_loan.annual_inc,2)
```

```python
### Convert dates to proper format as well as seperate them to columns
## Notably :
### issue_d
### earliest_cr_line
### last_pymnt_d
### last_credit_pull_d
### Are having Dates in format 'Mth-yy'. This can be intepreted in the date format by Python as format='%b-%Y'
filtered_loan.issue_d = pd.to_datetime(filtered_loan.issue_d, format='%b-%y')
filtered_loan.earliest_cr_line = pd.to_datetime(filtered_loan.earliest_cr_line, format='%b-%y')
filtered_loan.last_pymnt_d = pd.to_datetime(filtered_loan.last_pymnt_d, format='%b-%y')
filtered_loan.last_credit_pull_d = pd.to_datetime(filtered_loan.last_credit_pull_d, format='%b-%y')
```

```python
### Strip the strings from Data which can be categorised
## Notably :
### term : Has String Months
### int_rate : Has String %
### revol_util : Has String %
### emp_length : Has String <,>,years

### term
filtered_loan.term.unique()
filtered_loan.term = filtered_loan.term.str.replace(' months','')
### int_rate
filtered_loan.int_rate.unique()
filtered_loan.int_rate = filtered_loan.int_rate.str.replace('%','')
### revol_util
filtered_loan.revol_util.unique()
filtered_loan.revol_util = filtered_loan.revol_util.str.replace('%','')
### emp_length
filtered_loan.emp_length.unique()
emp_lengthMap = {'< 1 year' : 0, '1 year' : 1,  '2 years' : 2, '3 years' : 3, '4 years' : 4, '5 years' : 5, '6 years' : 6,  '7 years' : 7,  '8 years' : 8, '9 years' : 9, '10+ years' : 10
}
filtered_loan = filtered_loan.replace({"emp_length": emp_lengthMap })
filtered_loan[['term','int_rate','revol_util','emp_length']]
```

```python
## Copy the fine tuned Data to a new CSV File
filtered_loan.to_csv('final_loan.csv', index=False)
```

```python
# Data Analysis
# work with plotly
import plotly.offline as py 
py.init_notebook_mode(connected=True) # this code, allow us to work with offline plotly version
import plotly.graph_objs as go # it's like "plt" of matplot
import plotly.tools as tls # It's useful to we get some tools of plotly
import warnings # This library will be used to ignore some warnings
from collections import Counter # To do counter of some features
import plotly.io as pio
pio.renderers.default = "colab" #Colab configs for plotly
```

```python
##Import the final_loan.csv file via read_csv
final_loan = pd.read_csv('final_loan.csv',index_col=0,low_memory=False)
final_loan.shape
```

```python
## Understand the total count of defaulters and Non-Defaulters
tr0 = go.Bar(
            x = final_loan[final_loan["loan_status"]== 'Fully Paid']["loan_status"].value_counts().index.values,
            y = final_loan[final_loan["loan_status"]== 'Fully Paid']["loan_status"].value_counts().values,
            name='Not Defaulted'
    )

tr1 = go.Bar(
            x = final_loan[final_loan["loan_status"]== 'Charged Off']["loan_status"].value_counts().index.values,
            y = final_loan[final_loan["loan_status"]== 'Charged Off']["loan_status"].value_counts().values,
            name='Defaulted'
    )

data = [tr0, tr1]

layout = go.Layout(
    
)

layout = go.Layout(
    yaxis=dict(
        title='Count'
    ),
    xaxis=dict(
        title='Defaulter Variable'
    ),
    title='Loan Status distribution'
)

fig = go.Figure(data=data, layout=layout)

py.iplot(fig, filename='grouped-bar')
```

```python
##From the above graph, its evident that around 16 percent of the loan is charged off when compared with the fully paid Loans.
##Rise in this percentage over 20, would lead to drastic loss as per Industry standards
```

```python
## Funded Amount Vs Charged off Vs Non Defaulted Comparison
finalLoan_NonDefaulter = final_loan[final_loan["loan_status"] == 'Fully Paid']
finalLoan_Defaulter = final_loan[final_loan["loan_status"] == 'Charged Off']
tr0 = go.Box(
    y=finalLoan_NonDefaulter["funded_amnt"],
    x=finalLoan_NonDefaulter["loan_status"],
    name='Good credit',
    marker=dict(
        color='#3D9970'
    )
)

tr1 = go.Box(
    y=finalLoan_Defaulter['funded_amnt'],
    x=finalLoan_Defaulter['loan_status'],
    name='Bad credit',
    marker=dict(
        color='#FF4136'
    )
)
    
data = [tr0, tr1]

layout = go.Layout(
    yaxis=dict(
        title='Funded Amount',
        zeroline=False
    ),
    xaxis=dict(
        title='Funded Amount Wise Distribution'
    ),
    boxmode='group'
)
fig = go.Figure(data=data, layout=layout)

py.iplot(fig, filename='box-age-cat')
```

```python
## We Observe That most loan charged off are in the range of 6k to 16k. We must stricten the scrutiny of loans where request is in this particular range
```

```python
## Interest Rate Vs Charged off Vs Non Defaulted Comparison
finalLoan_NonDefaulter = final_loan.loc[final_loan["loan_status"] == 'Fully Paid']['int_rate'].values.tolist()
finalLoan_Defaulter = final_loan.loc[final_loan["loan_status"] == 'Charged Off']['int_rate'].values.tolist()

#First plot
tr0 = go.Histogram(
    x=finalLoan_NonDefaulter,
    histnorm='probability',
    name="Not Defaulted"
)
#Second plot
tr1 = go.Histogram(
    x=finalLoan_Defaulter,
    histnorm='probability',
    name="Defaulted"
)

#Creating the grid
fig = tls.make_subplots(rows=2, cols=2, specs=[[{}, {}], [{'colspan': 2}, None]],
                          subplot_titles=('Non Defaulted','Defaulted', 'In percentage'))

#setting the figs
fig.append_trace(tr0, 1, 1)
fig.append_trace(tr1, 1, 2)

fig['layout'].update(showlegend=True, title='Interest Rate Wise Distribuition', bargap=0.05)
py.iplot(fig, filename='custom-sized-subplot-with-subplot-titles')
```

```python
## It is clearly evident that for interest below 10%,  defaulting is low. It rises steep for interest above 10-16%.
##It is advisable to provide short term loan at lower interest rate than higher interest rate loans
```

```python
## Income Vs Charged off Vs Non Defaulted Comparison
incomePercentile = final_loan[final_loan.annual_inc < final_loan.annual_inc.quantile(0.95)]
finalLoan_NonDefaulter = incomePercentile.loc[incomePercentile["loan_status"] == 'Fully Paid']['annual_inc'].values.tolist()
finalLoan_Defaulter = incomePercentile.loc[incomePercentile["loan_status"] == 'Charged Off']['annual_inc'].values.tolist()

#First plot
tr0 = go.Histogram(
    x=finalLoan_NonDefaulter,
    histnorm='probability',
    name="Not Defaulted"
)
#Second plot
tr1 = go.Histogram(
    x=finalLoan_Defaulter,
    histnorm='probability',
    name="Defaulted"
)

#Creating the grid
fig = tls.make_subplots(rows=2, cols=2, specs=[[{}, {}], [{'colspan': 2}, None]],
                          subplot_titles=('Non Defaulted','Defaulted', 'General Distribuition'))

#setting the figs
fig.append_trace(tr0, 1, 1)
fig.append_trace(tr1, 1, 2)

fig['layout'].update(showlegend=True, title='Income Wise Distribuition', bargap=0.05)
py.iplot(fig, filename='custom-sized-subplot-with-subplot-titles')
```

```python
## We observe that borrowers with income ranging from 30k – 50k seems to have higher default rates. This needs to be evaluated as it can be due to higher number of loans been taken by this income group. This shall be evaluated in Bivariate Analysis
```

```python
## DTI Vs Charged off Vs Non Defaulted Comparison
finalLoan_NonDefaulter = final_loan.loc[final_loan["loan_status"] == 'Fully Paid']['dti'].values.tolist()
finalLoan_Defaulter = final_loan.loc[final_loan["loan_status"] == 'Charged Off']['dti'].values.tolist()

#First plot
tr0 = go.Histogram(
    x=finalLoan_NonDefaulter,
    histnorm='probability',
    name="Not Defaulted"
)
#Second plot
tr1 = go.Histogram(
    x=finalLoan_Defaulter,
    histnorm='probability',
    name="Defaulted"
)

#Creating the grid
fig = tls.make_subplots(rows=2, cols=2, specs=[[{}, {}], [{'colspan': 2}, None]],
                          subplot_titles=('Non Defaulted','Defaulted', 'General Distribuition'))

#setting the figs
fig.append_trace(tr0, 1, 1)
fig.append_trace(tr1, 1, 2)

fig['layout'].update(showlegend=True, title='DTI Wise Distribuition', bargap=0.05)
py.iplot(fig, filename='custom-sized-subplot-with-subplot-titles')
```
